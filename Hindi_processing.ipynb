```
def preprocess_hin(article):
    text = sub(r'[a-zA-Z]','', article) 
    text = sub(r'[0-9][0-9]', '', text)
    text = sub(r'\n', '', text)
    text = sub(r',','', text)
    text = sub(r':','', text)
    text = sub(r'[0-9]', '', text)
    text = sub(r'।', '', text)
    text = sub(r'-','', text)
    
    
    text = text.strip(string.punctuation)

    return text 
```

```
def tokenize_hin(article):
    nlp = Hindi()
    doc = nlp(article)
    tokens = [token.text for token in doc]
    
    return tokens 
```

```
#  getting all the tokens 
for i in range(len(news)):
    news[i] = preprocess_hin(news[i])
    news[i] = tokenize_hin(news[i])
news[100]
```

```
# making a global dictionary
global_dict = []
for i in range(len(news)):
    for word in news[i]:
        global_dict.append(word)
len(global_dict)
```

```
# finding all the unique words 
global_dict = list(set(global_dict))
len(global_dict)
```

```
# making the word vectors and storing them in a dictionary 
import gensim
model = gensim.models.KeyedVectors.load_word2vec_format('/content/drive/My Drive/jio_project/sentiment_analysis/sentiment_hi/cc.hi.300.vec')
```

```
word_map = []
num_words = 0
exceptions = []
for word in global_dict:
    try:
        word_dict = {}
        word_dict['word'] = word
        word_dict['vector'] = model[word]
        word_map.append(word_dict)
    except KeyError:
        print(word)
        exceptions.append(word)
        num_words = num_words +1 
print("=============================================================================")
print(num_words)
```

```
OUT_DIR = '/content/drive/My Drive/jio_project/sentiment_analysis/sentiment_hi/'
import json 
for i in exceptions:
    j = json.dumps(i)
    f = open(OUT_DIR+'exceptions_hindi.json','a')
    f.write(j)
    f.close()
    print(i," :saved")
```

```
for i in word_map:
    i['vector'] = i['vector'].tolist()
    j = json.dumps(i)
    f = open(OUT_DIR+'word_map_hindi.json', 'a')
    f.write(j)
    f.close()
    print(i['word'], " saved!")
```
