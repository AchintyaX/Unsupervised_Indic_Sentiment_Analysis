```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
def detokenize(article):
    tokens = [token for token in article]
    sentence = ' '.join(tokens)
    return sentence 
```

```
for i in range(len(news)):
    news[i] = detokenize(news[i])
```

```
tfidf = TfidfVectorizer()
tfidf.fit(news)
```

```
# making a global dictionary
global_dict = []
for i in range(len(news)):
    for word in news[i]:
        global_dict.append(word)
len(global_dict)
```

```
# finding all the unique words 
global_dict = list(set(global_dict))
len(global_dict)
```

```
doc_freq = {}
for i in range(len(news)):
    for j in global_dict:
        if j in doc_freq:
            continue
        else:
            doc_freq[j] = 0

        if j in news[i]:
            doc_freq[j] = doc_freq[j] + 1
```
